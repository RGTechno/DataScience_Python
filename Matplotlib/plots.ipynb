# **Different Plots in Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x+3

print(x,y1,y2)
```

```python
plt.style.available
```

```python
plt.style.use("seaborn")
```

### *1. Line Plot*

```python
plt.plot(x,y1,color="red",label="Bitcoin",marker="o")
plt.plot(x,y2,label="Ethereum",linestyle="dashed",marker="*")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Stock Price over Time")
plt.legend()
plt.show()
```

```python
# Note: We could also provide with only single 1D array as plot argument if X axis       data is not available.
# So it plots the array as Y axis and indices of arrays as X axis

plt.plot(y2)
plt.show()
```

### *2. Scatter Plot*

```python
# To adjust size of any plot
plt.figure(figsize=(4,4)) # Makes a 4x4 graph

plt.scatter(x,y1,color="red",label="Bitcoin",marker="o")
plt.scatter(x,y2,label="Ethereum",linestyle="dashed",marker="*")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot - Stock Price over Time")
plt.legend()
plt.show()
```

### *3. Bar Graph*

```python
x_cord = np.array([0,1,2])*2

heights = np.array([10,20,30])

plt.bar(x_cord-0.2,height=heights,width=0.5,label="Current Year",tick_label=["Gold","Platinum","Diamond"])
plt.bar(x_cord+0.2,height=heights*2+5,width=0.5,label="Previous Year")
plt.xlabel("Metal")
plt.ylabel("Price"),
plt.ylim(0,80)   # add limits to the y axis
plt.title("Bar Chart")
plt.legend()
plt.show()
```

### *4. Pie Charts*

```python
subject=["English","Maths","Physics","Chemistry","CS"]
weightage = [10,20,30,40,50]

plt.pie(weightage,labels=subject,explode=(0,0,0,0.1,0),shadow=True,autopct='%1.1f%%')
plt.show()
```

### *5. Histograms*

```python
## Standard Normal Distribution is where the mean = 0 and standard deviation = 1

# random.randn is one function in numpy which generate values in standard normal distribution

# To produce normal distribution(of mean u and std dev sigma) from standard normal distribution

# Xnd = Xsnd*sigma + u
```

```python
Xsnd = np.random.randn(100)

u=60
sigma = 5

Xnd = np.round(Xsnd*sigma+u)
Xnd1 = np.round(Xsnd*7+40)


print(Xnd)  # See Xnd has most values near 60

# print(Xsnd)
```

```python
plt.hist(Xnd,label="Maths",alpha=0.8)
plt.hist(Xnd1,label="Physics",alpha=0.9)

plt.xlabel("Marks")
plt.ylabel("No. of students")
plt.title("Histogram")

plt.legend()
plt.show()
```
