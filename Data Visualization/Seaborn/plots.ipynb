```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.barplot(x="sex", y="total_bill", data=tips,estimator=np.std)
## This is the aggregate feature plot

# Because of estimator now it shows the plot with aggregate standard deviation
```

```python
sns.countplot(x="sex",data=tips)

# This is the simple count plot signifies no of male and female in dataset as bar
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)

# Denotes max min with lines, horizontal represent mean
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="sex",split=True)

# Similar to boxplot
```

```python
sns.displot(tips["total_bill"],kind="kde")
```

```python
sns.kdeplot(tips["total_bill"])
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
```

```python
sns.pairplot(tips,hue="sex")
```
