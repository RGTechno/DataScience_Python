```python
import cv2 # Imports OpenCV Library

import matplotlib.pyplot as plt
```

```python
cv2.imread('../images/tbbt.jpg') 
#  Reads and Load image in a matrix format of 3 Dimension which are basically the RGB Channels.
#   R => Red
#   G => Green
#   B => Blue

# Note: OpenCV reads image in BGR format which is Blue,Green and Red Respectively
```

```python
img = cv2.imread('../images/tbbt.jpg')

plt.imshow(img)  # Displays Respective image

# Note: You might notice in output the color scheme is bit odd right? That's Because OpenCV Reads the image in BGR format and we need to convert in RGB or desired format/
```

```python
rgb_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # Converts BGR image to RGB image
```

```python
plt.imshow(rgb_image)
```

```python
plt.axis("off")
plt.imshow(rgb_image)
```

```python
# Cropping Image
```

```python
crop_img = rgb_image[500:1800,150:350,:]
plt.imshow(crop_img)
plt.axis("off")
```

```python
rgb_image.shape # Indicates the shape of image and dimension of Channels 2000x1465 in this case
```

```python
total_pixels = rgb_image.shape[0]*rgb_image.shape[1]
print(total_pixels*3)

# this implies when start to create real world data sets we need to resize the image because the original image takes up alot of int to store it. Hence inefficient
```

```python
# Resizing the image
```

```python
resized_img = cv2.resize(rgb_image,(200,150))
# .resize() method resizes the specified image to desired size 200x150 in this case
```

```python
plt.imshow(resized_img)
plt.title("TBBT")
```

```python
grey_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #Gray Image

plt.imshow(grey_img)
plt.axis("off")
```
