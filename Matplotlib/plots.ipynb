# **Different Plots in Matplotlib**

### *1. Line Plot*

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x+3

print(x,y1,y2)
```

```python
plt.style.available
```

```python
plt.style.use("seaborn")
```

```python
plt.plot(x,y1,color="red",label="Bitcoin",marker="o")
plt.plot(x,y2,label="Ethereum",linestyle="dashed",marker="*")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Stock Price over Time")
plt.legend()
plt.show()
```
