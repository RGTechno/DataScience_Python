```python
#Imports matplotlib library and called pyplot class with alias plt to access functions with plt
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = x**2+3 # Numpy array
```

```python
print(x,y)
```

```python
plt.plot(x,y) # Plots Line Chart with 1st and 2nd argument as x and y axis respectively
```

```python
plt.style.available # Gives available themes for graphs
```

```python
plt.style.use("seaborn")
```

```python
plt.plot(x,y,label="Stock 1") 
plt.plot(x,x**2,label="Stock 2")
# lable is used to lable the line with some colors(Legends)

plt.legend()  # It creates representative color legends to identify different LinePlot 

plt.xlabel("Time") # Labels x axis

plt.ylabel("Price") # Labels y axis

plt.title("Line Chart") # Gives a Title to entire Plot

plt.show() # Shows the Plot
```

```python
# .scatter method is used to plot a scatter point graph
plt.scatter(x,y,label="Stock 1") 
plt.scatter(x,x**2,label="Stock 2")
# lable is used to lable the line with some colors(Legends)

plt.legend()  # It creates representative color legends to identify different LinePlot 

plt.xlabel("Time") # Labels x axis

plt.ylabel("Price") # Labels y axis

plt.title("Scatter Plot") # Gives a Title to entire Plot

plt.show() # Shows the Plot
```
